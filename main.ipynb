```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from datetime import date

# Study data files
cities_path = "C:/Users/Sweety/Documents/GitHub/Web-Design-Challenge/Resources/cities.csv"

cities_df = pd.read_csv(cities_path)
##html = cities_df.to_html()
##print(html)
cities_df.describe()
```

```python
plt.scatter(cities_df["Lat"],cities_df["Max Temp"],edgecolors="grey",facecolors="red")
plt.title("Latitude vs. Max Temperature (2021-07-24)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (f)")
##plt.grid (b=False,which="major",axis="both",linestyle="-",color="lightgrey")
plt.grid(color='grey', linestyle='-.', linewidth=0.7)
plt.savefig("C:/Users/Sweety/Documents/GitHub/Web-Design-Challenge/Images/latitude_vs_temperature_ud_01.png")
plt.show()
```
